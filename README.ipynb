### Read a table from Gapminder
To get the URL of a table, pull the link from the 'Download' column and change the output from `xlsx` to `csv`.

```python
adults_with_hiv = Table.read_table('https://docs.google.com/spreadsheet/pub?key=pyj6tScZqmEfbZyl0qjbiRQ&output=csv',sep=',')
```

```python
adults_with_hiv
```

```python
# the country label is not so friendly, let's alias
country_column_label = adults_with_hiv.labels[0]
```

```python
top_1990 = adults_with_hiv.select(country_column_label, "1990").sort("1990", descending=True).where("1990", are.above(0))
```

```python
top_1990.barh(country_column_label)
```

## Compare 1990 and 2011

```python
top_1990_with_2011 = adults_with_hiv.select(country_column_label, "1990", "2011").sort("1990", descending=True).where("1990", are.above(0))
```

```python
top_1990_with_2011.barh(country_column_label)
```
